```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the outer donut (market share)
market_share = [62,11,6,5,16]  # Market share percentages

# Data for the inner donut (revenue percentage)
Revenue_in_2023 = [69.3	,13.3	,7.4	,7.1  ,14.7]  # Revenue percentages
marketCap = [877, 284 ,80]
colors1 = [
    '#003366',  # Navy Blue
    '#66B3FF',  # Sky Blue
    '#008080',  # Teal
    '#669966',  # Mint Green
    '#808080'   # Light Gray
]
# Create figure and axes
fig, ax = plt.subplots()
# Function to make text white for labels and autopct
def make_white_text(pcts):
    for text in pcts:
        text.set_color('white')

# First (outer) donut chart (market share)
wedge1, texts, autopct1 = ax.pie(market_share, colors=colors1, labels=["Market Share", "", "", "",""], autopct='%1.1f%%', startangle=100, 
                                 wedgeprops=dict(width=0.3),  # Donut shape
                                 pctdistance=0.85, textprops={'color': 'white', 'fontweight': 'bold'})  # Distance of values from the center


# Make the outer pie percentage values white
make_white_text(autopct1)

colors2 = [
    '#003366',  # Navy Blue
    '#66B3FF',  # Sky Blue
    '#808080'   # Light Gray
]

# Second (inner) donut chart (revenue)
ax2 = plt.subplot(111, frameon=False)
wedges2, texts2, autopct2 = ax2.pie(marketCap, colors=colors2, 
                                     autopct='%1.1f%%', startangle=100, 
                                     wedgeprops=dict(width=0.5),
                                     pctdistance=0.8, textprops={'color': 'white', 'fontweight': 'bold'})  # Base position for labels inside the donut

# Adjust the position of the inner donut
ax2.set_position([0.263, 0.245, 0.5, 0.5])

# Manually adjust the positions of labels in the inner donut
label_positions = [(0.5, 0.1), (0.5, 0.1), (0.5, 0.1), (0.5, 0.1),(0.5, 0.1),(0.5, 0.1)]  # Adjust these values as needed
for i, label in enumerate(texts2):
    # Update the position of each label manually
    x, y = label_positions[i]
    label.set_position((x, y))

# Add a centered text for revenue
ax2.text(0, 0, 'Total Revenue\n$112 Billion', ha='center', va='center', fontsize=8, fontweight='bold')  # Replace $X with actual value


# Define additional information for each company
extra_info = ['TSMC Revenue\n$69.3B', 'Samsung\n Revenue\n$13.3B', 'GlobalFoundries\nRevenue\n$7.4B',
              'UMC\nRevenue\n$7.1B', 'Others Revenue\n$15.5B']

# Loop through each company to add connecting lines and colored boxes
for i, (info, wedge) in enumerate(zip(extra_info, wedge1)):
    # Calculate the mid-angle of each wedge
    angle = (100 + (sum(market_share[:i]) + market_share[i] / 2) / sum(market_share) * 360)
    
    # Convert angle to radians and calculate x, y for the wedge center point
    x = np.cos(np.radians(angle))
    y = np.sin(np.radians(angle))
    
    # Annotate each company with additional info and a color box
    ax.annotate(info, xy=(x * 1, y * 1),  # Position on the pie chart
                xytext=(x * 1.5, y * 1.5),  # Position for the label
                arrowprops=dict(facecolor='black', arrowstyle="->", lw=1.5),  # Arrow style
                bbox=dict(boxstyle="round,pad=0.5", edgecolor="black", facecolor=wedge.get_facecolor(), lw=2),  # Box color matches wedge
                color='white',fontweight = 'bold',  # Set the text color to white
                ha='center', fontsize=10)

ax.annotate('Outer pie = Market Share\n Inner pie = Market Capitalization', xy=(x * 1.0, y * 1.0),  # Position at the outer surface of the pie chart
            xytext=(x * 2, y * 2),  # Position for the label farther away
              # Arrow style
            bbox=dict(boxstyle="round,pad=0.5", edgecolor="black", facecolor='white', lw=2),  # White box for the label
            ha='center', fontsize=10, fontweight='bold', color='black')  # Bold black text

# Add a comment at the bottom of the chart
plt.text(0.5, -0.3, 'This analysis belongs to Pkeday Advisors (Source: Statista)', ha='left', fontsize=8, transform=plt.gca().transAxes)
# Show the plot
plt.show()
```
